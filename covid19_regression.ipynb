```python
# menghapus kolom yang terlalu banyak null nya
df = df.drop(['City or Regency', 'Special Status'],axis=1)
```

```python
# menghapus data yang nilaiiya kosong
df = df.dropna(axis=0)
df.isnull().sum()
```

```python
# import seaborn as sns
# corrmat = df.corr()
# sns.heatmap(corrmat, annot=True)
```

```python
# melakukan reset pada index
df = df.reset_index(drop=True)
```

```python
# mengubah CFR menjadi integer
df_CRF = df['Case Fatality Rate'].values

CRF = []
for i in df_CRF:
    CRF.append(float(i[:-1]))

# mengubah CRR menjadi integer
df_CRR = df['Case Recovered Rate'].values

CRR = []
for i in df_CRR:
    CRR.append(float(i[:-1]))

df['Case Fatality Rate'] = CRF
df['Case Recovered Rate'] = CRR
```

```python
# Transformasi Data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

fitur = ['New Cases', 'New Deaths', 'New Recovered', 'New Active Cases', 'Total Cases', 'Total Deaths','Total Recovered', 'Total Active Cases', 'Total Regencies', 'Total Cities', 'Total Districts', 'Total Urban Villages', 'Total Rural Villages', 'Area (km2)', 'Population', 'Population Density','New Cases per Million', 'Total Cases per Million', 'New Deaths per Million', 'Total Deaths per Million', 'Case Fatality Rate', 'Case Recovered Rate', 'Growth Factor of New Cases','Growth Factor of New Deaths']
df_fitur = df[fitur]
data_scaled = scaler.fit_transform(df_fitur)
column_name = list(df_fitur.columns)
df_fitur = pd.DataFrame(data=data_scaled, columns=column_name)
df_fitur
```

```python
for i in fitur:
    df[i] = df_fitur[i]
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

x = df_fitur.drop(['New Cases', 'New Active Cases'], axis=1)
y = [int(i) for i in df_fitur['New Cases']]
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(x, y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Nama Fitur','Score']  #naming the dataframe columns
print(featureScores.nlargest(77,'Score'))  #print 10 best features
```

```python

from sklearn.linear_model import (
    LinearRegression, TheilSenRegressor, RANSACRegressor, HuberRegressor)
model2 = make_pipeline(PolynomialFeatures(4), LinearRegression())
model3 = make_pipeline(PolynomialFeatures(4), TheilSenRegressor())
model4 = make_pipeline(PolynomialFeatures(4), RANSACRegressor())
model5 = make_pipeline(PolynomialFeatures(4), HuberRegressor())

model2.fit(X_train, y_train)
model3.fit(X_train, y_train)
model4.fit(X_train, y_train)
model5.fit(X_train, y_train)

data = {
    'x': np.array([i[0] for i in X_train]),
    'y': y_train.values,
}

apakah = pd.DataFrame(data)
apakah = apakah.sort_values(by='x')

x_poly = np.array(apakah['x']).reshape(-1,1)

y_ridge2 = model2.predict(x_poly)
y_ridge3 = model3.predict(x_poly)
y_ridge4 = model4.predict(x_poly)
y_ridge5 = model5.predict(x_poly)

# plt.scatter(X_train, y_train)


plt.legend(('model2', 'model3', 'model4'),loc='lower right')

plt.scatter(X_train, y_train)
plt.plot(x_poly, y_ridge2, color='gold',)
plt.plot(x_poly, y_ridge3, color='red',)
plt.plot(x_poly, y_ridge4, color='green',)
plt.plot(x_poly, y_ridge5, color='purple',)
plt.show()
```

```python
print('Regresi Linear \t\t',model2.score(X_test, y_test))
print('TheilSenRegressor \t',model3.score(X_test, y_test))
print('RANSACRegressor \t',model4.score(X_test, y_test))
print('Huber Regressor \t',model5.score(X_test, y_test))
```
